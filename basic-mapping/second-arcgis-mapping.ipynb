  - and **`Feature Classes`**

<mark>The relationship between a geodata base, a feature dataset, and feature dataset is that it is like a hiarchy system down the ladder. The geodatabase is the thing that holds all of the information for the project including feature datasets and feature classes. It is like a large folder (as seen in the image above) Databases take all of this complex information and compress it in a neat, low volume file for your computer. The feature dataset, contained in the database is a collection of related feature classes. This is like a folder contained in the first folder (as seen in the image above) This compilation of feature classes is used to create large parts of the map such as the topology. Lastly, a feature class is a actual individual data layer. A map/ geodatabase contains many of these in order to make up the whole picture.  </mark>
